## Exemplo de regress√£o linear utilizando o Polyfit 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy.polynomial.polynomial import polyfit
california_housing = pd.read_csv('/content/sample_data/california_housing_test.csv')
plt.figure(figsize=(10,8))

median_income = california_housing['median_income']

median_house_value = california_housing['median_house_value']

plt.scatter(median_income, median_house_value)

x = median_income

y = median_house_value

b, m = polyfit(x, y, 1)

reg = b + m * x


plt.scatter(x, y, c="b", alpha=0.1, marker=r'o', label="Data points", s=0.5)
plt.plot(x, reg, '-', c='r')

```

```python
#Trabalhando com dados
import matplotlib.pyplot as plt

x = [0.1, 1.1, 1.8, 2.7, 2.9, 3.0, 3.1, 3.2]

y = [0.3, 0.6, 1.2, 2.5, 2.9, 3.1, 3.2, 3.3]

plt.axis([0, 4, 0, 4])

#Plotando os dados
plt.plot(x, y, label='Linha')
plt.ylabel('legenda eixo y')
plt.xlabel('legenda eixo x')
plt.legend(loc='upper left')
plt.title('Título')
plt.show()
```

* **plt.xlabel() e plt.ylabel()**: Usado para definir legendas do eixo.
* **plt.title()**: Usado para definir o título do gráfico.

* **plt.legend(loc='  ')**: Usado para definir a localização das legendas do gráfico.  'upper right',
'upper left',  'lower right',  'lower left', 'best'

    
    Obs.: Para definir as legendas do gráfico, é necessário criar [ Label="  "] na funçao plt.plot().

```python

import matplotlib.pyplot as plt

x = [0.1, 1.1, 1.8, 2.7, 2.9, 3.0, 3.1, 3.2]

y = [0.3, 0.6, 1.2, 2.5, 2.9, 3.1, 3.2, 3.3]


#Configurando os eixos
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.axis([0,4,0,4])

#Personalizando as linhas dos eixos 
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.spines['top'].set_linewidth(1.5)

#Personalizando o gráfico 
plt.plot(x, y, linewidth=2.0, c='navy', marker='o', markersize=5.0, mfc='black', ls='-.', label="linha")
plt.legend(loc='best')
plt.xlabel('Eixo X', fontsize=10, fontweight= 'bold' )
plt.ylabel('Eixo Y', fontsize=10, fontweight= 'bold' )
plt.title('Gráfico', fontsize=15, color='black', fontname='sans-serif', fontweight= 'bold' )
plt.grid(True)
plt.show()
```



* **ax.spines['  '].setlinewidth(' ')**: Definir a espessura da linha do gráfico.


    obs.1: ['  '] = selecionar a linha -----> 'right', 'left', 'bottom', 'top'.
   
    obs.2: ('  ') = definir espessura. (Float)

    
* **plt.grid(True)**: Adicionar linhas de grade.
   

```python
import matplotlib.pyplot as plt
import numpy as np


precos = np.arange(0,50,1)

def oferta(precos):
  return 20+6*precos

def demanda(precos):
  return 250-(5*precos)

#Plotando a função
plt.figure(figsize= (8,5))
plt.axis([0,400,0,60])
plt.plot(demanda(precos),precos, c="b", lw=2.0, label='Demand')
plt.plot(oferta(precos),precos, c="r", lw=2.0, label='Supply')


#Traçando os pontos de equilíbrio 

plt.plot([145,0], [21, 21], '--')
plt.plot([145,145], [21, 0], '--')
plt.scatter(145, 21, c='black')

#
plt.xlabel('Quantity', fontsize=10, fontweight= 'bold' )
plt.ylabel('Price', fontsize=10, fontweight= 'bold' )
plt.legend(loc='best')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.title('Supply x Demand', fontsize=15, color='black', fontname='sans-serif', fontweight= 'bold' )
plt.show()
```

```python
from matplotlib import pyplot as plt


year = [2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013,
        2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]  

growth_rate = [0.12, 0.24, 0.26, 0.30, 0.40, 0.43, 1.0, 1.5,
               2.0, 2.1, 2.2, 2.4, 2.6, 2.8, 2.5, 2.3, 2.0,
               2.6, 2.9, 3.0, 3.3, 3.5, 3.6]

growth_rate_2 = [0.1, 0.21, 1.8, 2.8, 2.45, 2.41, 2.2, 2.8,
               3.3, 3.5, 4.0, 4.2, 4.3, 4.0, 4.2, 4.3, 4.8,
               5.0, 5.1, 5.2, 5.5, 5.6,6.0]

#Criando dois plots "ax1" e "ax2".
fig, (ax1, ax2) = plt.subplots(2, 1, sharey=True, figsize = (9,7))

#Ajuste de espaçamento entre os plots.
fig.subplots_adjust(hspace=0.3)

#Plotando os dados 
ax1.plot(year, growth_rate, alpha=0.5, marker="o", color="b")
ax2.plot(year, growth_rate_2, alpha=0.5, marker="o", color="r")

#Definindo as legendas para "ax1" e "ax2"
ax1.set_ylabel("Growth rate 1")
ax2.set_ylabel('Growth rate 2')
ax2.set_xlabel('Time (years)')
ax1.set_title("Population growth rate", fontsize=15)
plt.show()
```

## Scatter Plots (Gráficos de dispersão) --- plt.scatter

```python
import numpy as np
import matplotlib.pyplot as plt


year = ([2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013,
        2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024])

growth_rate = ([0.12, 0.24, 2.0, 3.0, 0.40, 0.43, 1.0, 1.5,
               2.0, 3.1, 1.3, 2.4, 4.6, 2.8, 2.5, 4.3, 5.8,
               2.6, 1.9, 3.0, 3.5, 5.3, 3.6])

#Podemos chamar as variavéis 'year' e 'growth_rate' de 'x' e 'y'.

x = year

y = growth_rate

#Definindo o tamanho do plot
plt.figure(figsize=(8,5))

#Utilizamos a função plt.scatter para chamar os gráficos de disperção.
plt.scatter(x, y, c="r", alpha=0.7, marker=r'o',
            label="Data points")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title('Scatter Plot',fontsize=15, color='black')
plt.legend(loc='upper left')
plt.show()
```

## Bar Plot - Gráfico de barras


```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()


países = ('País 1', 'País 2', 'País 3', 'País 4', 'País 5')
produção = (2, 4, 6, 3, 5)


ax.bar(países, produção, align='center', color='b', width=0.5)
ax.set_title('Produção de aço em toneladas')

plt.show()
```
